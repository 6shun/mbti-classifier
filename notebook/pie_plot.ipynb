```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
root_path = '/content/drive/MyDrive/CS289A Project/'
data_path = root_path + 'data/'
train_df_path = data_path + 'mbti_1.csv'
```

```python
df = pd.read_csv(train_df_path)
```

```python
df = df.groupby('type').count().rename(columns={'Unnamed: 0' :'Count'})
df.reset_index(inplace=True)
df
```

# Pie chart of mbti type

```python
plt.figure(figsize=(12,6))
plt.pie(df.posts, labels=df.type, autopct='%1.1f%%')
plt.legend()
plt.axis('equal')
plt.tight_layout()
```

```python
plot_labels = ['ENFJ',
 'ENFP',
 'ENTJ',
 'ENTP',
 '',
 '',
 '',
 '',
 'INFJ',
 'INFP',
 'INTJ',
 'INTP',
 'ISFJ',
 'ISFP',
 'ISTJ',
 'ISTP']
```

```python
x = df['type']
y = df['posts']
porcent = 100.*y/y.sum()

plt.figure(figsize=(12,8))
patches, texts = plt.pie(y,labels=plot_labels, startangle=0, radius=1)
labels = ['{0} - {1:1.1f} %'.format(i,j) for i,j in zip(x, porcent)]

sort_legend = True

if sort_legend:
    patches, labels, dummy =  zip(*sorted(zip(patches, labels, y),
                                          key=lambda x: x[2],
                                          reverse=True))
plt.legend(patches, labels, loc=8, bbox_to_anchor=(0.9, 0.2, 0.5, 0.5),
           fontsize=11)
```
